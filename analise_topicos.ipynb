```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
import matplotlib.dates as mdates
```

```python
data = pd.read_csv('speeches_topics.csv')
data.head()
```

```python
# Certifique-se de que 'year' está no formato correto
data['year'] = pd.to_datetime(data['date']).dt.year

# Preparar dados para LSA
topic_count_lsa = data.groupby(['year', 'dominant_topic_lsa']).size().unstack(fill_value=0)
topic_count_lsa = topic_count_lsa.reindex(columns=["inflation", "activity", "economic crisis", "monetary policy"], fill_value=0)

# Preparar dados para LDA
topic_count_lda = data.groupby(['year', 'dominant_topic_lda']).size().unstack(fill_value=0)
topic_count_lda = topic_count_lda.reindex(columns=["inflation", "activity", "economic crisis", "monetary policy"], fill_value=0)

# Configurando a figura para conter dois subplots (um para LSA e outro para LDA)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6), sharey=True)

# Plotando as séries temporais para o modelo LSA
topic_count_lsa.plot(ax=axes[0], marker='o')
axes[0].set_title('Séries Temporais para LSA')
axes[0].set_xlabel('Ano')
axes[0].set_ylabel('Contagem de Tópicos')
axes[0].grid(True)

# Plotando as séries temporais para o modelo LDA
topic_count_lda.plot(ax=axes[1], marker='o')
axes[1].set_title('Séries Temporais para LDA')
axes[1].set_xlabel('Ano')
axes[1].set_ylabel('Contagem de Tópicos')
axes[1].grid(True)

# Adicionar legendas e ajustar o layout
fig.legend(["inflation", "activity", "economic crisis", "monetary policy"], loc='upper right')
plt.tight_layout()
plt.show()
```
