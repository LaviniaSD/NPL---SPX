```python
import pandas as pd
import numpy as np

data = pd.read_csv("all_ECB_speeches.csv", sep="|", encoding="utf-8")
print("Shape DataFrame: ", data.shape)
print("Data size: ", data.shape[0])
data_drop_nan = data.dropna(subset=data.columns)
data.dropna(subset=data.columns, inplace=True)
print("Dropando NaN")
print("Data size: ", data_drop_nan.shape[0] )
data.head()
```

```python
#um loop que mostra um content e espera imput do usuário que devolve A, B ou C. o valor é guardado numa lista. o loop é interrompido se o usuário não concede um valor. a lista é inserida como coluna

import pandas as pd

# create a dataframe with sample data
df = pd.DataFrame({'content': ['content1', 'content2', 'content3']})
df = data

# create an empty list to store user inputs
user_inputs = []

# loop through each row in the dataframe
for index, row in df.iterrows():
    a = 0
    b = 0
    c = 0
    d = 0
    content = row['contents']
    paragraphs = content.split("\n")
    for paragraph in paragraphs:
        sentences = paragraph.split(".")
        for sentence in sentences:
            print(sentence)
            user_input = input(f"Enter a value for '{sentence}': ")
            
            if user_input == "A":
                a += 1
            elif user_input == "B":
                b += 1
            elif user_input == "C":
                c += 1
            else:
                d += 1
    user_input = input(f"Enter a value for speech given A={a}, B={b}, C={c}, D={d}: ")
    
    # check if user input is empty
    if not user_input:
        break
    
    # check if user input is valid
    if user_input not in ['A', 'B', 'C']:
        print("Invalid input. Please enter 'A', 'B', or 'C'.")
        continue
    
    # append user input to the list
    user_inputs.append(user_input)

# add the user_inputs list as a new column in the dataframe
df['user_input'] = user_inputs

# print the updated dataframe
print(df)
```

```python
#pra cada um em speakers, quero seccionar o content em paragrafos, depois frases. Sorteie três frases de cada speaker e guarde 

import random

def select_sentences(data):
    # Create a dictionary to store the selected sentences for each speaker
    selected_sentences = {}

    # Loop through each speaker in the data
    for speaker in data['speakers'].unique():
        # Get the content for the current speaker
        content = data[data['speakers'] == speaker]['contents'].values[0]
        
        # Split the content into paragraphs
        paragraphs = content.split("\n")
        
        # Initialize a list to store the selected sentences for the current speaker
        speaker_sentences = []
        
        # Loop through each paragraph
        for paragraph in paragraphs:
            # Split the paragraph into sentences
            sentences = paragraph.split(".")
            
            # Remove empty sentences
            sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
            
            # Randomly select three sentences
            selected = random.sample(sentences, min(10, len(sentences)))
            
            # Add the selected sentences to the list
            speaker_sentences.extend(selected)
        
        # Store the selected sentences for the current speaker in the dictionary
        selected_sentences[speaker] = speaker_sentences

    # Print the selected sentences for each speaker
    for speaker, sentences in selected_sentences.items():
        print(f"Speaker: {speaker}")
        print("Selected Sentences:")
        for sentence in sentences:
            print(sentence)
        print()

        # Specify the file path
        file_path = "selected_sentences_2.txt"

        # Open the file in write mode
        with open(file_path, "w") as file:
            # Write the selected sentences for each speaker to the file
            for speaker, sentences in selected_sentences.items():
                file.write(f"Speaker: {speaker}\n")
                file.write("Selected Sentences:\n")
                for sentence in sentences:
                    file.write(sentence + "\n\n")
                file.write("\n")

        # Print a message indicating that the sentences have been saved
        print("Selected sentences have been saved to the file:", file_path)

select_sentences(data)

#monetary policy, activity, inflation, other
# , employment, financial stability, climate change, digitalization, inequality, globalization, trade, fiscal policy, structural reforms, central bank communication, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability, financial crisis, financial cycle, financial system, financial sector, financial services, financial instruments, financial markets, financial institutions, financial regulation, financial supervision, financial intermediation, financial innovation, financial inclusion, financial literacy, financial technology, financial stability,
```

```python
# o arquivo teve as linhas brancas preenchidas. recupere os dados relacionando cada frase ao que está na linha seguinte
def collect_sentence_label_pairs(selected_sentences, labels):
    sentence_label_pairs = []
    
    for speaker, sentences in selected_sentences.items():
        for sentence in sentences:
            user_input = input(f"Enter a label for the sentence: '{sentence}': ")
            
            # check if user input is valid
            if user_input not in labels:
                print("Invalid label. Please enter a valid label.")
                continue
            
            # append the sentence and label pair to the list
            sentence_label_pairs.append((sentence, user_input))
    
    return sentence_label_pairs

# Specify the file path
file_path = "selected_sentences_2.txt"

# Read the selected sentences from the file
pairs = collect_sentence_label_pairs(file_path)

#crie um dataframe com as frases e as categorias
labeldata = pd.DataFrame(columns=["sentence", "category"])
for pair in pairs:
    labeldata = labeldata.append({"sentence": pair[0], "category": pair[1]}, ignore_index=True)
print(labeldata.head())
```
